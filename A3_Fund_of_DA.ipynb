# Take appropriate actions to handle the outliers.

```python
# Specify the attributes for which you want to remove outliers
attributes_to_remove_outliers = ["DAYS_EMPLOYED", "CNT_FAM_MEMBERS", "DAYS_REGISTRATION", "OBS_60_CNT_SOCIAL_CIRCLE"]

def remove_outliers_iqr_specific(df_filled, attributes_to_remove_outliers):
    df_filtered = df_filled.copy()  # Make a copy of the original DataFrame to avoid modifying it directly
    for attribute in attributes_to_remove_outliers:
        Q1 = df_filtered[attribute].quantile(0.25)
        Q3 = df_filtered[attribute].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_filtered = df_filtered[(df_filtered[attribute] >= lower_bound) & (df_filtered[attribute] <= upper_bound)]
    return df_filtered


# Remove outliers for the specified attributes and update the filtered DataFrame
df_filtered = remove_outliers_iqr_specific(df_filled, attributes_to_remove_outliers)

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(13, 4 * num_rows))
fig.suptitle("Box Plots of Numerical Attributes (After Removing Outliers)", fontsize=16)

# Iterate over numerical attributes and create subplots
for i, attr in enumerate(numerical_attributes):
    row_idx = i // 3
    col_idx = i % 3

    ax = axes[row_idx, col_idx] if num_attributes > 1 else axes  # Handle single attribute case
    sns.boxplot(data=df_filtered[attr], ax=ax, color='lightblue')
    ax.set_title(f'Box Plot for {attr}')
    ax.set_ylabel(attr)

# Remove any empty subplots (if num_attributes is not a multiple of 3)
for i in range(num_attributes, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.subplots_adjust(top=0.75)  # Adjust the title's position
plt.show()
```

# **UP SAMPLING**

```python
import pandas as pd
from sklearn.utils import resample

df_up = df_filled.copy()

# Separate the majority (class 0) and minority (class 1) classes
majority_class = df_up[df_up['TARGET'] == 0]
minority_class = df_up[df_up['TARGET'] == 1]

# Upsample the minority class to match the majority class
minority_upsampled = resample(minority_class,
                              replace=True,  # Sample with replacement
                              n_samples=len(majority_class),  # Match the majority class
                              random_state=500)  # For reproducible results

# Combine the majority and upsampled minority classes
df_up = pd.concat([majority_class, minority_upsampled])
```
